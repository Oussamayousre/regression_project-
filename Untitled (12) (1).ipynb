```r
library(ggplot2)
```

```r
# La régression linéaire R utilise la fonction lm () pour créer un modèle de régression à partir d'une formule, sous la forme de Y ~ X + X2.
ozone <- read.table("projet.txt", header=TRUE, sep=";", dec=",")

reg_multi <- lm(Y~X1+X2+X3,data=ozone)
AIC(reg_multi)
BIC(reg_multi)

summary(reg_multi)
```

```r
#On va donc maintenant retirer la variable non significative. qui est X2  car elle a une p-valeur de 0.528.(p-value< 0.05 )
reg_multii <- lm(Y~X1+X3,data=ozone)
AIC(reg_multii)
BIC(reg_multii)

summary(reg_multii)
```

```r
#En statistique, le coefficient de détermination, noté R2 ou r2, est une mesure de la qualité de la prédiction d'une régression linéaire.
#on a  R-squared = 0.7024 Cela implique que 70.24% de la variabilité de la variable dépendante a été prise en compte et que les 29.76% restants de la variabilité sont toujours non pris en compte.
X1<-ozone$X1   #on affecte X1 au valeur X1 de projet.txt
Y<-ozone$Y    
plot(X1,Y) #pour présenter les graphes 
ggplot(ozone,aes(x=X1,y=Y))+
  geom_point()+
  stat_smooth(method="lm",se=FALSE)+
  xlab("X1")+
  ylab("Y")
#on observe que le nuage des points nous fait penser à un alignement selon une forme qui n'est pas très loin d'une droite.
```

```r
X2<-ozone$X2
plot(X2,Y) 
ggplot(ozone,aes(x=X2,y=Y))+
  geom_point()+
  stat_smooth(method="lm",se=FALSE)+
  xlab("X2")+
  ylab("Y")
#on observe que Y ne peut étre expliquer par X2 (random values)
```

```r
X3<-ozone$X3
plot(X3,Y)
ggplot(ozone,aes(x=X3,y=Y))+
  geom_point()+
  stat_smooth(method="lm",se=FALSE)+
  xlab("X3")+
  ylab("Y")
#on observe que le nuage des points nous fait penser à un alignement selon une forme qui n'est pas très loin d'une droite.
```

```r
#Si l'on souhaite prévoir la concentration journalière en ozone, sachant que la température sera de 40 °C, que la valeur de  X3 sera de 53, 
predit <- data.frame(X1=40, X3=53)
Y_prev <- predict(reg_multi,predit)
round(Y_prev, digits=2)
#la valuer prédit de Y ( maximum de concentration d'ozone observé sur la journée) est 140.15 
```
